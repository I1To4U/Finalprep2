```python
#Question No 2
balloons = {"red": 5, "yellow": 10, "green": 7}

while True: #I use loop to let you choose more than 1 time
    balloon_color = input("Enter the color of the balloon to get a free balloon (or type 'exit' to quit): ")

    if balloon_color.lower() == "exit":
        print("Thank You.")
        break

    if balloon_color in balloons:
        if balloons[balloon_color] > 0:
            balloons[balloon_color] -= 1
            print("You got the free balloon.")
        else:
            print("Color out of stock.")
    else:
        print("Color not found.")
```

```python
#Question No 3
def calculate_shipping_cost(weight_kg):

    if weight_kg <= 5:
        shipping_cost = weight_kg*6
    elif weight_kg <= 20:
        shipping_cost = weight_kg*4
    else:
        shipping_cost = weight_kg*2
    return shipping_cost

weight_kg = float(input("Enter your package's weight."))
cost_amount = calculate_shipping_cost( weight_kg)
print("Total shipping cost =", cost_amount)
```
